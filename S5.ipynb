```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
import utils
from model import Net
```

```python
device = utils.get_device()
```

```python
train_loader, test_loader = utils.get_train_test_data()
```

```python
batch_data, batch_label = next(iter(train_loader))
utils.plot_images(batch_data, batch_label)
```

```python
train_losses = []
train_acc = []
test_losses = []
test_acc = []
```

CODE BLOCK: 10

```python
model = Net().to(device)
optimizer = utils.get_optimizer(model)
scheduler = utils.scheduler(optimizer)
# New Line
criterion = utils.loss_function()
num_epochs = 20
for epoch in range(1, num_epochs+1):
  print(f'Epoch {epoch}')
  acc, loss=utils.train(model, device, train_loader, optimizer, criterion)
  train_losses.append(loss)
  train_acc.append(acc)
  acc, loss = utils.test(model, device, train_loader, criterion)
  test_losses.append(loss)
  test_acc.append(acc)
  scheduler.step()
```

CODE BLOCK: 11

```python
utils.plot_loss_acc_curves(train_losses, train_acc, test_losses, test_acc)
```
